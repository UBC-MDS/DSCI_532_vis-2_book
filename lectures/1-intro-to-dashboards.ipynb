## 1. What is a **data dashboard**?  

 "An **information management tool** that visually tracks, analyzes, and displays key performance indicators, metrics, as well as key data points, allowing you to monitor the current state of your business, department, team, or specific process"
[Source: Databox](https://databox.com/what-is-a-dashboard#what)


### 1.1 What technology we can use to create one?

There are many popular technologies available to create dashboards for data scientists. Tableau, Streamlite, Voila, Shiny, Dash, Power BI, are some of the most common ones.

A Shiny app is a web page (UI) connected to a computer running a live R session (Server)


Shiny Apps have 2 components:

- **User Interface (UI)**
Define how your app look. It is also called *frontend*

- **Server**
Define how your app works. It is also called *backend*

- `shinyApp()`
Connects the UI and the server
